# Business Understanding.
## Overview.
Syria tel is Telecommunications company that offers mobile services, including voice, text, and data plans to its customers. Hence the company is intrested in understanding factors that influence customer retention and identifying customers who are at risk of `churn`.

**Churn analysis** this is the process of examining why customers stop using a product or service, helping businesses identify factors leading to customer attrition and develop strategies to improve retention.
Through churn analysis Syriatel can proactively target at-risk customers with retention strategies and improve overall customer satisfaction.

## Problem Statement.
Syriatel faces a challenge in understanding key factors that contribute to customer churn. The company extensively collected customer data, however they still lack the ability to analyse and predict customer churn accurately. Through churn analysis and Machine learning algorithms the company will be able to understand, analyse and predict this will ultimately improve business insights and prevent cutomer attrition.

## Objectives.
### Main Objectives.
- Predict probability of customer churn with SyriaTel telecommunications.

### Specific Objectives.
- Analyse how long a customer has had an account in SyriaTel.
- Understand how customers use their plan by calculating the average call duration.
- Analyse the number of customer service calls.
- Voice Mail Plan vs Voicemail Messages- This is to analyze whether customers with a voicemail plan actually use it.
- International plan usage.


### Key metrics to focus on
Main Metric: `Recall.`
- This measures how well the model identifies the positives (the churners)
- Ensure a high Recall score this will indicate that there are few missed churners.

Other metrics:
- Accuracy - This measures the percentage of model's correct predictions. 
- Precision - Mesures how many predicted churners are actually churners.
- F1-score - Balances the Recall and precision.

# Data Understanding
The dataset was sourced from Kaggle. It has 3333 features and 21 columns. The information contained in the dataset will give insights on the probability of a customer to churn.
The data set comprises of:
- state: Customer's geographic location.

- account length: Duration (in months) the customer has been with Syriatel.

- area code: Numeric code for the customer's geographic region.

- phone number: Customer’s phone number.

- international plan: Whether the customer has an international calling plan.

- voice mail plan: Whether the customer has a voicemail service.

- number vmail messages: Number of voicemail messages received by the customer.

- total day minutes: Total minutes used by the customer during the day.

- total day calls: Total calls made by the customer during the day.

- total day charge: Charges for the customer's day usage.

- total eve minutes: Total minutes used by the customer during the evening.

- total eve calls: Total calls made by the customer during the evening.

- total eve charge: Charges for the customer's evening usage.

- total night minutes: Total minutes used by the customer during the night.

- total night calls: Total calls made by the customer during the night.

- total night charge: Charges for the customer's night usage.

- total intl minutes: Total minutes used by the customer for international calls.

- total intl calls: Total number of international calls made by the customer.

- total intl charge: Charges for the customer's international calls.

- customer service calls: Number of calls made to customer service.

- churn: Whether the customer has churned (left the service) or not.

### Feature Engineering
- Feature Engineering is the process of transforming raw data into meaningful features that can be used for analysis and increase model accuracy.

```python
data.head()
```

- Here columns that contain sensitive information will be dropped.

```python
# drop columns 
data= data.drop(columns= ['area code', 'phone number'])
data.columns
```

#### Categorize Account Length

```python
# check values in account length
data['account length'].unique()
```

```python
# categorize the column 
def categorize_values(val):
    if val <= 81:
        return 'short-term'
    if 82 <= val <= 162:
        return 'mid-term'
    else:
        return 'long-term'
    
# apply function to the column
data['account_duration']= data['account length'].apply(categorize_values)
data.tail(3)
```

#### Average Daily Usage.

✅ This columns will be identifies customers who make frequent short calls vs few long calls.

Thus answering the question. ➡️ On average, how long is each call the customer makes?

```python
# calculate total calls 
data['total_calls'] = data['total day calls'] + data['total eve calls'] + data['total night calls']

# calculate total minutes
data['total_min'] = data['total day minutes'] + data['total eve minutes'] + data['total night minutes']

# calculate the average daily use of a customer
data['avg_usage']= data['total_min'] / data['total_calls']

data.head(3)
```

## Exploratory Data Analysis.
### Univariate Analysis.
#### Identify percentage of Customers who have churned.

```python
# count of churn distribution
churn_count = df1['churn'].value_counts()

# create a piplot to show thw distribution
plt.figure(figsize=(8,6))
plt.pie(churn_count, labels=churn_count.index, autopct='%1.1f%%', startangle=140, colors=["Green", 'lightgreen'])

# create a blank circle at the centre
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# customize the plot
plt.title("Distribution of Customer Churn")
plt.legend(["1= Active", "0= Churned"], loc="upper right")
plt.show()
```
